### Exponential spatial distribution

$ f(x) = \frac{1}{X_0} e^{-x/X_0}$

```python
X0 = 8.5
x  = np.power(10.0,np.arange(0.0,1.5,0.01))
plt.plot( x, np.exp(-x/X0)/X0 )
plt.title( 'Example Probability Density Distribution' )
```

Integrate to get the cumulative distribution function, then set equal to $y$ (picked randomly from 0-1)

$F(x) = 1 - e^{-x/X_0} = y$

```python
plt.plot( x, 1-np.exp(-x/X0) )
plt.title( 'Example Cumulative Density Distribution' )
```

$x = -X_0 \ln( 1 - y )$

```python
NSAMP = 1000
Y_R = np.random.random(NSAMP)
Y_Z = np.random.random(NSAMP)
```
