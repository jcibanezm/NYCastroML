<h2>Random numbers?</h2>

Computers have some tricks to produce truly random numbers. For example, some Linux systems will time keystrokes or measure internal voltage variations to generate a pool of constantly updating random numbers, but often you'll do fine with _pseudorandom_ numbers. 

What is a Pseudorandom number generator? Well, Wikipedia tells us that a function $f$ is a Pseudorandom number generator iff:

A pseudorandom number generator produces numbers that look statistically random, but were created in an entirely deterministic manner. You can __seed__ such generators to always produce the same sequence of numbers. For example, this code will always produce the same sequence of numbers (even on different machines):

```python
import random
random.seed(42)
[random.random() for ii in range(5)]
```

Most pseudorandom number generators generate uniformly distributed samples. For example, one of the earliest algorithms was introduced by von Neumann (actually, originally by a Friar in the 1200's...) called the _Middle-square method_. 

The idea is simple:

 * Start from a seed number with $n$ digits
 * Square the number
 * Take the middle $n$ digits to use as the new seed

```python
def middle_square(seed, nsamples=1):
    x = seed**2
    
    # empty list to append samples to
    samples = []
    for ii in range(nsamples):
        # select the middle four digits for the next seed
        newseed = (x / 100) % 10000
        x = newseed**2
        samples.append(newseed)
    
    return samples
```

```python
print middle_square(2349, 10)
```

Well, those look pretty random! But wait, what if we run for a little bit longer...

```python
print middle_square(2349, 100)
```

Notice that it gets trapped in a cycle: `4100, 8100, 6100, 2100` etc.

Ok, well, there are much better algorithms, but many have some inherent (very, very long) periodicity to them. So just keep that in mind!

<h2>Sampling from arbitrary distributions</h2>

Let's now assume we can generate statistically uniform random numbers. How do we sample from other distributions? I won't talk about many methods that (hopefully?) are covered later in the book, but I think it's useful to think about this.

One very simple way to generate samples from arbitrary distributions is with __rejection sampling__. The concept is perhaps best visualized below, in an example that computes the value of $\pi$:

 * generate two sets of uniform random numbers $x\in\mathcal{U}(0,1)$, $y\in\mathcal{U}(0,1)$,
 * compute the distance from the origin for all points $(x,y)$,
 * select all points that lie within $r<1$
 * compute the ratio num. $r<1$ / total
 * this ratio $\times~4 \approx \pi$

```python
nsamples = 10000
x = np.random.uniform(0.,1.,size=nsamples)
y = np.random.uniform(0.,1.,size=nsamples)
r = np.sqrt(x**2 + y**2)

sum(r < 1) / float(nsamples) * 4
```

```python
fig,ax = subplots(1,1,figsize=(5,5))

# plot the unit circle
c = Circle((0.,0.), radius=1, color='#333333')
ax.add_patch(c)

# plot the data points
ax.plot(x[r<=1], y[r<=1], marker='.', 
        linestyle='none', color='#1a9641', alpha=0.5)
ax.plot(x[r>1], y[r>1], marker='.',
        linestyle='none', color='#d7191c', alpha=0.7)

ax.set_xlim(0,1)
ax.set_ylim(0,1)
```

Here's another example of how you could sample from a Gaussian...but you shouldn't do this! Why? You either have to restrict to some domain (meaning you're not strictly sampling from a Gaussian anymore), or you choose an infinitely large domain and end up trying to sample from a delta function.

If we can write out the functional form for the pdf, we can use rejection sampling to sample from some domain:

```python
def normal(x, mu, var):
    return np.exp(-0.5*(x-mu)**2/var) / np.sqrt(2*np.pi*var)
```

The procedure here is slightly different, but the same idea. We essentially draw a box around the function, sample from the box, accept any point with y < f(x), then collect the x values:

```python
nsamples = 10000
mu = 5.
variance = 1.

x = np.random.uniform(0, 10, size=nsamples)
y = np.random.uniform(0, (2*np.pi*variance)**-0.5, size=nsamples)
```

```python
func_x = np.linspace(0., 10.)
plot(func_x, normal(func_x, mu, variance), 
     linewidth=2., color='#333333')

idx = y <= normal(x, mu, variance)
plot(x[idx], y[idx], marker='.', 
     linestyle='none', color='#1a9641', alpha=0.5)
plot(x[~idx], y[~idx], marker='.', 
     linestyle='none', color='#d7191c', alpha=0.5)

yt = yticks([])
```

Again, we:

* sample x and y uniformly over some box around the function of interest,
* reject any points where y > f(x),
* take the accepted x values, and now we have values distributed according to the target distribution
