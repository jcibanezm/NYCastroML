```python
import numpy as np
```

```python
def monty_hall(switch_choice, niter=10000, ndoors=3):
    """ Simulate the Monty Hall problem.
        
        Parameters
        ----------
        switch_choice : bool
            Switch to one of the doors left after the host 
            discards one, or keep initial choice.
        niter : int
            Number of iterations to do. Should be large if you
            want a meaningful result.
        ndoors : int
            Number of doors to simulate.
    """
            
    nwins = 0
    for ii in range(niter):
        doors = np.zeros(ndoors, dtype=int)
        doors[np.random.randint(ndoors)] = 1
        
        choice = np.random.randint(ndoors)
        if doors[choice] == 1:
            # if the contestant gets lucky
            nwins += 1
            continue
        
        w, = np.where(doors == 0)
        host_doors = np.delete(doors, w[np.random.randint(ndoors-1)])
        
        if switch_choice:
            # switch to a randomly chosen door from the subsample chosen by the host
            choice = np.random.randint(ndoors-1)
        else:
            # keep the same choice
            pass
        
        if doors[choice] == 1:
            nwins += 1

    return nwins / float(niter)
```

```python
all_ndoors = np.arange(3, 10, 1)

switcher = []
stubborn = []
for ndoors in all_ndoors:
    stubborn.append(monty_hall(False, ndoors=ndoors))
    switcher.append(monty_hall(True, ndoors=ndoors))
```

```python
figure(figsize=(10,5))
       
plot(all_ndoors, stubborn, label="Stays with first choice", lw=2.)
plot(all_ndoors, switcher, label="Switches", lw=2.)

xlabel("Number of doors")
ylabel("Chance of winning")

legend()
```
