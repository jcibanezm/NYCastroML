# Radial and vertical distribution of stars.

I'll do the IMF later, right now I want to draw the radial and heigh distribution of stars.

I want to draw a pseudo-random number, between [ 0.0,1 ) and map it into an exponential distribution function.

\begin{equation} f(x) = \frac{1}{\it{x}_{0}} \it{e}^{-x/\it{x}_{0}} \end{equation}

```python
# Number of stars that I want to sample.
Nstars = 1.0e4
```

```python
# Let's sample the radial distance from 0.0 to 100.0 kpc every 1 kpc.
x = np.arange(0.0,100,1.0)
x0 = 10.0
fx = Nstars/x0*np.exp(-x/x0)
plt.plot(x,fx, 'b')
plt.title('Radial distribution of stars on the disk galaxy')
plt.xlabel('Radial distance [kpc]')
plt.ylabel('Number of stars')
```

Now I want to plot the cumulative density distribution.

$F(x) = \int_{0}^{\infty} f(x) dx$

This should add up to the normalization constant, i.e. Total number of stars

```python
Fx = Nstars*(1.0-np.exp(-x/x0))
plt.plot(x,Fx, 'r')
plt.title('Cumulative density distribution')
```

# For the Mass Function

the universal Salpeter initial mass function has the form: 
$$ \frac{dN}{dm} = k m^{-\alpha} $$
where k is a normalization constant and $\alpha = 2.35$

Integrating the IMF we obtain the total number of stars:

$$N(m) = \int_{m_{0}}^{m_{1}} k m ^{-\alpha} dm $$

with the limits 
    $ m_{0} = 0.1 M_{\odot} $ and 
    $ m_{1} = 100.0 M_{\odot} $

```python
alpha = 2.35
m0 = 0.1
m1 = 100.0
```

Where the Cumulative density function is given by
$$1 = F(x) = \int_{m_{0}}^{m_{1}}km^{-\alpha}dm $$

and the normalization constant given by this cumulative distribution is:

$$ k = \frac{1-\alpha}{m^{1-\alpha}-m_{0}^{1-\alpha}}$$

```python
# Cumulative distribution function constant.
kCumul = (1.0 - alpha) / (m1**(1.0 - alpha) - m0**(1.0 - alpha))
```

Inverting the cumulative distribution function we obtain the mapping function of random numbers into a Salpeter IMF,
i.e. the mass generating function from a uniforma random number:

$$ m(X) = [ X(m) \frac{1-\alpha}{k}+ M_{0}^{1-\alpha}]^{1/1-\alpha} $$

where m(X) is the mass generating function from the random number X.
and X(m) is a pseudorandom number between [0.0 , 1]

```python
Xm = np.random.random( Nstars*1000 )
```

```python
mX = ( Xm*(1.0-alpha)/kCumul + m0**(1.0-alpha))**(1.0/(1.0-alpha))
```

```python
b = plt.hist(np.log10(mX), log=True, bins=100)
plt.xlabel('Mass [$M_{\odot}$]')
plt.ylabel('Number of stars N(m)')
```

```python
Mtot = np.sum(mX)
print "Total mass in stars in the Galaxy", Mtot
```
