# Section 4.5. Confidence Estimates: the Bootstrap and the Jacknife

## Bootstrapping

Bootstrapping is a computationally fast way to estimate the uncertainty of a certain statstic when used on your data, instead of on the (unknown) underlying PDF. We collect data to study the underlying PDF, but we cannot "know" what that PDF is, so we cannot describe it with population statistics. But we do "know" our data, so we can use them as a proxy of the PDF in order to study the different statistics that we use.

Assume we want to know the average size of cats' tails. There are only so many *N* cats that we can catch and measure safely in a given amount of time.
In Bootstrap, we create *B* new samples by _resampling with replacement_ from the original sample.
Each of these *B* samples yields one measure of the statistic we're after, $\mu$.
A histogram of these measures will show how the statistic $\mu$ varies, based on our data.

This is the *non-parametric* version of bootstrap. The *parametric* version draws samples from the best-fitting model instead of from the data.

Note: because there are $N!$ distinct sets of our $N$ measurements, for large $N$, the number of possible samples quickly becomes $>>1$. This means that, practically speaking, each time we run bootstrap, we will get slightly different results. I assume (but don't know) that the variance between results will usually be smaller than the intrinsic variance in the statistic we're testing.

```python
r"""
Bootstrap Calculations of Error on Mean
---------------------------------------
Figure 4.3.

The bootstrap uncertainty estimates for the sample standard deviation
:math:`\sigma` (dashed line; see eq. 3.32) and :math:`\sigma_G` (solid line;
see eq. 3.36). The sample consists of N = 1000 values drawn from a Gaussian
distribution with :math:`\mu = 0` and :math:`\sigma = 1`. The bootstrap
estimates are based on 10,000 samples. The thin lines show Gaussians with
the widths determined as :math:`s / \sqrt{2(N - 1)}` (eq. 3.35) for
:math:`\sigma` and :math:`1.06 s / \sqrt{N}` (eq. 3.37) for :math:`\sigma_G`.
"""
# Author: Jake VanderPlas
# License: BSD
#   The figure produced by this code is published in the textbook
#   "Statistics, Data Mining, and Machine Learning in Astronomy" (2013)
#   For more information, see http://astroML.github.com
#   To report a bug or issue, use the following forum:
#    https://groups.google.com/forum/#!forum/astroml-general

from scipy.stats import norm
from astroML.resample import bootstrap
from astroML.stats import sigmaG

#----------------------------------------------------------------------
# This function adjusts matplotlib settings for a uniform feel in the textbook.
# Note that with usetex=True, fonts are rendered with LaTeX.  This may
# result in an error if LaTeX is not installed on your system.  In that case,
# you can set usetex to False.

m = 1000  # number of points
n = 10000  # number of bootstraps

#------------------------------------------------------------
# sample values from a normal distribution
np.random.seed(123)
data = norm(0, 1).rvs(m)

#------------------------------------------------------------
# Compute bootstrap resamplings of data
mu1_bootstrap = bootstrap(data, n,  np.std, kwargs=dict(axis=1, ddof=1))
mu2_bootstrap = bootstrap(data, n, sigmaG, kwargs=dict(axis=1))

#------------------------------------------------------------
# Compute the theoretical expectations for the two distributions
x = np.linspace(0.8, 1.2, 1000)

sigma1 = 1. / np.sqrt(2 * (m - 1))
pdf1 = norm(1, sigma1).pdf(x)

sigma2 = 1.06 / np.sqrt(m)
pdf2 = norm(1, sigma2).pdf(x)

#------------------------------------------------------------
# Plot the results
fig, ax = plt.subplots(figsize=(10, 10))

ax.hist(mu1_bootstrap, bins=50, normed=True, histtype='step',
        color='blue', ls='dashed', label=r'$\sigma\ {\rm (std. dev.)}$')
ax.plot(x, pdf1, color='gray')

ax.hist(mu2_bootstrap, bins=50, normed=True, histtype='step',
        color='red', label=r'$\sigma_G\ {\rm (quartile)}$')
ax.plot(x, pdf2, color='gray')

ax.set_xlim(0.82, 1.18)

ax.set_xlabel(r'$\sigma$')
ax.set_ylabel(r'$p(\sigma|x,I)$')

ax.legend()

plt.show()
```

## Jacknife

Similar to bootstrap, but instead of resampling with replacement, jacknife creates new samples by drawing subsamples of the data with $K < N$ data points.

For the special case where $K=N-1$, the bias-corrected jacknife estimate of our statistic, $\alpha^J$, is:

$$ \alpha^J = \alpha_N + \Delta(\alpha),$$

where $\alpha_N$ is the estimate of the statistic from our entire data sample and $\Delta(\alpha)$ is the bias correction:

$$\Delta(\alpha) = (N-1)\left(\alpha_N -\frac{1}{N}\sum_{i=1}^N \alpha_i^* \right)$$,

where $\alpha_i^*$ are the estimates of the statistic from each new subsample $i$.

Jacknife with $K=N-1$ is unreliable for statistics that are insensitive to outliers, such as the median or percentiles. For such statistics, it gains reliability as more data points are removed from the new samples. 

Note #1: whereas bootstrap gives slightly different results each time, jacknife gives only one result, as it uses all possible subsamples.

Note #2: both bootstrap and jacknife must be used with caution for small $N$.

Note #3: I tried to recreate the calculation below, using Matlab and no black boxes. I get $\sigma_G = 0.93 \pm 0.05$, not $0.60 \pm 0.03$.


```python
r"""
Jackknife Calculations of Error on Mean
---------------------------------------
Figure 4.4.

The jackknife uncertainty estimates for the width of a Gaussian distribution.
This example uses the same data as figure 4.3. The upper panel shows a
histogram of the widths determined using the sample standard deviation, and
using the interquartile range. The lower panel shows the corrected jackknife
estimates (eqs. 4.33 and 4.35) for the two methods. The gray lines show the
theoretical results, given by eq. 3.35 for :math:`\sigma` and eq. 3.37 for
:math:`\sigma_G`. The result for :math:`\sigma` matches the theoretical
result almost exactly, but note the failure of the jackknife to correctly
estimate :math:`\sigma_G` (see the text for a discussion of this result).
"""
# Author: Jake VanderPlas
# License: BSD
#   The figure produced by this code is published in the textbook
#   "Statistics, Data Mining, and Machine Learning in Astronomy" (2013)
#   For more information, see http://astroML.github.com
#   To report a bug or issue, use the following forum:
#    https://groups.google.com/forum/#!forum/astroml-general

#------------------------------------------------------------
# sample values from a normal distribution
np.random.seed(123)
m = 1000  # number of points
data = norm(0, 1).rvs(m)

#------------------------------------------------------------
# Compute jackknife resamplings of data
from astroML.resample import jackknife
from astroML.stats import sigmaG

# mu1 is the mean of the standard-deviation-based width
mu1, sigma_mu1, mu1_raw = jackknife(data, np.std,
                                    kwargs=dict(axis=1, ddof=1),
                                    return_raw_distribution=True)

pdf1_theory = norm(1, 1. / np.sqrt(2 * (m - 1)))
pdf1_jackknife = norm(mu1, sigma_mu1)

# mu2 is the mean of the interquartile-based width
#  WARNING: do not use the following in practice.  This example
#           shows that jackknife fails for rank-based statistics.
mu2, sigma_mu2, mu2_raw = jackknife(data, sigmaG,
                                    kwargs=dict(axis=1),
                                    return_raw_distribution=True)
pdf2_theory = norm(data.std(), 1.06 / np.sqrt(m))
pdf2_jackknife = norm(mu2, sigma_mu2)
print mu2, sigma_mu2

#------------------------------------------------------------
# plot the results
print "mu_1 mean: %.2f +- %.2f" % (mu1, sigma_mu1)
print "mu_2 mean: %.2f +- %.2f" % (mu2, sigma_mu2)

fig = plt.figure(figsize=(15, 8))
fig.subplots_adjust(left=0.11, right=0.95, bottom=0.2, top=0.9,
                    wspace=0.25)

ax = fig.add_subplot(121)
ax.hist(mu1_raw, np.linspace(0.996, 1.008, 100),
        label=r'$\sigma^*\ {\rm (std.\ dev.)}$',
        histtype='stepfilled', fc='white', normed=False)
ax.hist(mu2_raw, np.linspace(0.996, 1.008, 100),
        label=r'$\sigma_G^*\ {\rm (quartile)}$',
        histtype='stepfilled', fc='gray', normed=False)
ax.legend(loc='upper left', handlelength=2)

ax.xaxis.set_major_locator(plt.MultipleLocator(0.004))
ax.set_xlabel(r'$\sigma^*$')
ax.set_ylabel(r'$N(\sigma^*)$')
ax.set_xlim(0.998, 1.008)
ax.set_ylim(0, 550)

ax = fig.add_subplot(122)
x = np.linspace(0.45, 1.15, 1000)
ax.plot(x, pdf1_jackknife.pdf(x),
        color='blue', ls='dashed', label=r'$\sigma\ {\rm (std.\ dev.)}$',
        zorder=2)
ax.plot(x, pdf1_theory.pdf(x), color='gray', zorder=1)
ax.plot(x, pdf2_jackknife.pdf(x),
        color='red', label=r'$\sigma_G\ {\rm (quartile)}$', zorder=2)
ax.plot(x, pdf2_theory.pdf(x), color='gray', zorder=1)
plt.legend(loc='upper left', handlelength=2)

ax.set_xlabel(r'$\sigma$')
ax.set_ylabel(r'$p(\sigma|x,I)$')
ax.set_xlim(0.45, 1.15)
ax.set_ylim(0, 24)

plt.show()
```

# Section 4.6. Hypothesis Testing

## Definitions

**Null hypothesis**: the hypothesis we're looking to reject through our measurements. Using data, we can reject a null hypothesis, but we cannot prove it, since non-rejection might be caused by too few, or low-quality, data.

**p-value**: the probability that we would get a value at least as large as $x_i$ is $p(x>x_i) = 1-H(x_i)$, where $H(x)$ is the CDF of $h(x)$.

**significance level $\alpha$**: a threshold $p$-value used to reject the null hypothesis when $p \le \alpha$ (e.g., if we set $\alpha=0.05$ and find that $p<0.05$, then the null hypothesis is rejected at a 0.05, or 95%, significane level). The significance level has to be set *before* testing the null hypothesis.

**Type I error / False positive**: rejecting the null hypothesis, even though it is true. I.e., a spurious source that's really just background. The false positive probability is limited by the significance level $\alpha$.

**Type II error / False negative**: not rejecting the null hypothesis, even though it is false. I.e., failing to detect a real source in an image. The false negative probability is usually called $\beta$ and is related to *the power of a test* as $(1-\beta)$.

## The trade-off between completeness and contamination in simple classification

In every problem, there is a trade-off between the number of false positives and false negatives that is acceptable. This trade-off is set by the significance level $\alpha$, so that each problem has an optimal $\alpha$ value.

I many classification problems, where we have only two outcomes (either these are the doids we're looking for, or not), let $h_B(x)$ be our null hypothesis (not THE doids), and $h_S(x)$ be the PDF of the population of interest (droids on a mission). In more familiar terms, $h_B(x)$ is the background in an image, and $h_S(x)$ is the PDF of source counts in an image.

The population from which a single datum $x_i$ is drawn is a combination of the two populations, such that $p_B(x_i) = 1-p_S(x_i)$: 

$$h(x) = (1-a)h_B(x) + ah_S(x),$$

where $a$ is the relative normalization factor between the two PDFS, and $\int h_B(x)dx = \int h_S(x)dx = 1$.

If the sample size is $N$, then the number of real sources in the image is $Na$.

We have to choose a classification boundary $x_c$ so that if $x<x_c$, it is considered background, and if $x>x_c$, it is considered a source.

```python
"""
Example of classification
-------------------------
Figure 4.5.

An example of a simple classification problem between two Gaussian
distributions. Given a value of x, we need to assign that measurement to one
of the two distributions (background vs. source). The cut at xc = 120 leads
to very few Type II errors (i.e., false negatives: points from the distribution
hS with x < xc being classified as background), but this comes at the cost of
a significant number of Type I errors (i.e., false positives: points from the
distribution :math:`h_B` with x > xc being classified as sources).
"""
# Author: Jake VanderPlas
# License: BSD
#   The figure produced by this code is published in the textbook
#   "Statistics, Data Mining, and Machine Learning in Astronomy" (2013)
#   For more information, see http://astroML.github.com
#   To report a bug or issue, use the following forum:
#    https://groups.google.com/forum/#!forum/astroml-general

#------------------------------------------------------------
# Generate and draw the curves
x = np.linspace(50, 200, 1000)
p1 = 0.9 * norm(100, 10).pdf(x)
p2 = 0.1 * norm(150, 12).pdf(x)

fig, ax = plt.subplots(figsize=(10, 7.5))
ax.fill(x, p1, ec='k', fc='#AAAAAA', alpha=0.5)
ax.fill(x, p2, '-k', fc='#AAAAAA', alpha=0.5)

ax.plot([120, 120], [0.0, 0.04], '--k')

ax.text(100, 0.036, r'$h_B(x)$', ha='center', va='bottom')
ax.text(150, 0.0035, r'$h_S(x)$', ha='center', va='bottom')
ax.text(122, 0.039, r'$x_c=120$', ha='left', va='top')
ax.text(125, 0.01, r'$(x > x_c\ {\rm classified\ as\ sources})$')

ax.set_xlim(50, 200)
ax.set_ylim(0, 0.04)

ax.set_xlabel('$x$')
ax.set_ylabel('$p(x)$')
plt.show()
```

The *expected number of spurious sources* (false positives) is:

$$n_{spurious} = N(1-a)\alpha = N(1-a)\int_{x_c}^{\infty} h_B(x) dx,$$

and the number of missed sources (false negatives) is:

$$n_{missed} = Na\beta = Na\int_0^{x_c} h_S(x) dx.$$

The number of instances classified as a sourse (i.e., the null hypothesis is rejected) is:

$$n_{source} = Na - n_{missed} + n_{spurious}.$$

The sample *completeness* (sensitivity / recall rate) is defined as:

$$\eta = \frac{Na-n_{missed}}{Na} = 1 - \int_0^{x_c} h_S(x)dx, 0 \le \eta \le 1$$

The sample *contamination* is defined as:

$$\epsilon = \frac{n_{spurious}}{n_{source}}, 0 \le \epsilon \le 1.$$

$(1-\epsilon)$ is sometimes called the %classification efficinecy*.

As $x_c$ goes up, the contmination goes down, but so does the completeness ($n_{missed}$ goes up). How to find the optimal $x_c$?

## The Benjamini-Hochberg method for finding the optimal $x_c$

### Assumptions:

1. $h(x) = (1-a)h_B(x) + ah_S(x)$ <br><br>

2. h_B(x) is known (e.g., we can estimate the background of an image by)

### Method

- Assign a p-value to each datum, $p(x_i) = 1 - H_B(x_i)$. <br><br>

- Sort the sample of p-values by increasing $p_i$. <br><br>

If all $x_i$ were drawn from $h_B(x)$, then the cumulative distribution of p-values, $C$, would increase linearly as $C_i^B = N p_i$. <br><br>

But, if there are $Na$ sources in the data, then there will be an overexcess of small p-values, so that the real cumulative distribution $C(p_i)$ will have large values at small $p_i$. <br><br>

- Assume that the null hypothesis is rejected for all $p_i \le p_c$, so that the first $i_c$ values are selected as sources, corresponding to $Na$. <br><br>

The number of false positives is $(1-a)Np_c < Np_c$, so the contamination rate is: <br><br>

$$\epsilon = \frac{(1-a)Np_c}{i_c} < \frac{Np_c}{i_c}.$$

Which leads to the requirement that: 

$$i_c < N\frac{p_c}{\epsilon_0},$$

where $\epsilon_0$ is set by the user.

- So, for a given $\epsilon_0$, we find the largest $i_c$ that satisfies $i_c < N\frac{p_c}{\epsilon_0}$. $p_c$ is the significance level $\alpha$ we adopt.


# Section 4.7. Comparison of Distributions

## 4.7.1. Regression toward the Mean

If a random variable is extreme on its first measurement, it will tend to be closer to the population mean on a second measurement.

So, a statistical test applied to a *selected* subsample of the data might produce a different outcome than when applied to a *randomly selected* subsample.

Example: a new kind of test is administered to school kids, but only the scores of the bottom 25% of students are examined. Whether or not the test made any difference, if these students took the exam again, they would have different scores (becaues of random errors), closer to those of the mean population.

## 4.7.2. Nonparametric Methods for Comparing Distributions

Nonparametric = the underlying distributions are unknown.

**Kolmogorov-Smirnov (K-S) test**: compares the CDFs of two samples of data by measuring the distance between them. <br>
Rejection expectation: $D >> 1/\sqrt{N}$ <br>

Variations: <br><br> 
-. two-sample: compare two samples of data: "are the two data samples drawn from the same distribution?" <br>

-. one-sample: compare one sample of data to a known PDF: "is the data consistent with being drawn from this known distribution?"

Notes: <br><br>
-. More sensitive to the center of the distribution that to its tails, so not the best option for distinguishing samples drawn from Gaussian or exponential distributions.

-. Sensitive to more than one property of the underlying distribution.

Python routines: *kstest*, *ks\_2samp*, and *ksone* within scipy.stats.

See also: Cramer-von Mises criterion, Watson test, Anderson-Darling test.
The latter is more sensitive to the tails of the distribution

**The U test (Mann-Whitney-Wilcoxon / Wilcoxon rank-sum)**: compares the _medians_ of two samples of data.

**The Wilcoxon signed-rank test**: compares the _means_ of two samples of data.

Python routines: _mannwhitneyu_ and _wilcoxon_ in scipy.stats.

## 4.7.4. Comparing to a Gaussian distribution

Different statistical tests probe different parameters of the Gaussian's shape. A test that is sensitive to the mean/median will be sensitive to skewness, but not to kurtosis.

**Anderson-Darling test**: varies according to whether $\mu$ and $\sigma$ are determined from data. <br>
Rejection expectation: $A^2 >> 1$ <br>
Python routines: _anderson_ in scipy.stats.

```python
from scipy import stats
x = np.random.normal(0, 1, size=1000)
A, crit, sig = stats.anderson(x, 'norm')
A
```

In this case, both $\mu$ and $\sigma$ were not determined from the data, so the p-values for significance levels of 0.05 and 0.01 are 0.79 and 1.09, respectively. We get 0.33, which means that we cannot reject the null hypothesis that the data are consistent with a Gaussian distribution. 

**Shapiro-Wilk test**: very sensitive to non-Gaussian tails (i.e., outliers), but not very much to the core of the distribution.

Rejection expectation: $W < 1$

```python
stats.shapiro(x)
```

**$Z_1$ and $Z_2$ non-Gaussianity tests**

When $N>100$, the fraction $s/\sigma_G$ follows a nearly Gaussian distribution with $\mu\sim1$ and $\sigma\sim0.92/\sqrt{N}$.
Similary, the fraction $|\mu-q_{50}|/s$ follows a nearly Gaussian distribution with $\mu\sim0$ and $\sigma\sim0.76s/\sqrt{N}$.

$$Z_1 = 1.3 \frac{|\mu-q_{50}|}{s} \sqrt{N}$$

$$Z_2 = 1.1 \left|\frac{s}{\sigma_G}-1\right|\sqrt{N}$$

Rejection expectation: $Z_1$ and $Z_2$ are larger than several $\sigma$.

```python
"""
Gaussianity Tests
-----------------
Figure 4.7.

The results of the Anderson-Darling test, the Kolmogorov-Smirnov test, and the
Shapiro-Wilk test when applied to a sample of 10,000 values drawn from a normal
distribution (upper panel) and from a combination of two Gaussian distributions
(lower panel).

The functions are available in the ``scipy`` package:

- The Anderson-Darling test (``scipy.stats.anderson``)
- The Kolmogorov-Smirnov test (``scipy.stats.kstest``)
- The Shapiro-Wilk test (``scipy.stats.shapiro``)
"""
# Author: Jake VanderPlas
# License: BSD
#   The figure produced by this code is published in the textbook
#   "Statistics, Data Mining, and Machine Learning in Astronomy" (2013)
#   For more information, see http://astroML.github.com
#   To report a bug or issue, use the following forum:
#    https://groups.google.com/forum/#!forum/astroml-general

from astroML.stats import mean_sigma, median_sigmaG

# create some distributions
np.random.seed(1)
normal_vals = stats.norm(loc=0, scale=1).rvs(10000)
dual_vals = stats.norm(0, 1).rvs(10000)
dual_vals[:4000] = stats.norm(loc=3, scale=2).rvs(4000)

x = np.linspace(-4, 10, 1000)
normal_pdf = stats.norm(0, 1).pdf(x)
dual_pdf = 0.6 * stats.norm(0, 1).pdf(x) + 0.4 * stats.norm(3, 2).pdf(x)

vals = [normal_vals, dual_vals]
pdf = [normal_pdf, dual_pdf]
xlims = [(-4, 4), (-4, 10)]


#------------------------------------------------------------
# Compute the statistics and plot the results
fig = plt.figure(figsize=(10, 14))
fig.subplots_adjust(left=0.13, right=0.95,
                    bottom=0.06, top=0.95,
                    hspace=0.1)

for i in range(2):
    ax = fig.add_subplot(2, 1, 1 + i)  # 2 x 1 subplot

    # compute some statistics
    A2, sig, crit = stats.anderson(vals[i])
    D, pD = stats.kstest(vals[i], "norm")
    W, pW = stats.shapiro(vals[i])

    mu, sigma = mean_sigma(vals[i], ddof=1)
    median, sigmaG = median_sigmaG(vals[i])

    N = len(vals[i])
    Z1 = 1.3 * abs(mu - median) / sigma * np.sqrt(N)
    Z2 = 1.1 * abs(sigma / sigmaG - 1) * np.sqrt(N)

    print 70 * '_'
    print "  Kolmogorov-Smirnov test: D = %.2g  p = %.2g" % (D, pD)
    print "  Anderson-Darling test: A^2 = %.2g" % A2
    print "    significance  | critical value "
    print "    --------------|----------------"
    for j in range(len(sig)):
        print "    %.2f          | %.1f%%" % (sig[j], crit[j])
    print "  Shapiro-Wilk test: W = %.2g p = %.2g" % (W, pW)
    print "  Z_1 = %.1f" % Z1
    print "  Z_2 = %.1f" % Z2

    # plot a histogram
    ax.hist(vals[i], bins=50, normed=True, histtype='stepfilled', alpha=0.5)
    ax.plot(x, pdf[i], '-k')
    ax.set_xlim(xlims[i])

    # print information on the plot
    info = "Anderson-Darling: $A^2 = %.2f$\n" % A2
    info += "Kolmogorov-Smirnov: $D = %.2g$\n" % D
    info += "Shapiro-Wilk: $W = %.2g$\n" % W
    info += "$Z_1 = %.1f$\n$Z_2 = %.1f$" % (Z1, Z2)
    ax.text(0.97, 0.97, info,
            ha='right', va='top', transform=ax.transAxes)

    if i == 0:
        ax.set_ylim(0, 0.55)
    else:
        ax.set_ylim(0, 0.35)
        ax.set_xlabel('$x$')

    ax.set_ylabel('$p(x)$')

plt.show()
```
