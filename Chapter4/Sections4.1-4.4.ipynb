**Frequentist mindset**

+ Probabilities refer to the <i>frequencies of events</i> which are based on objective properties of the real world, which have <i>fixed</i> values, albeit unknown.<br>
+ A 95% confidence interval contains the true value of the parameter <i>95% of the time</i>.

**Bayesian mindset**

+ Probabilities refer to our <i>belief</i> in everything from data point measurements to model parameters.
+ We 95% believe that a model parameter falls within our 95% confidence interval, as gauged from the <i>distribution quantifying the uncertainty of our knowledge about the parameter</i> (the posterior distribution).

Fortunately, the arguments one might make about how to <b>interpret</b> results is subtle and you don't need to worry much about it... unless you're trying to win an argument with some statisticians.$^3$

# 4.2: Maximum Likelihood Estimation (MLE)

The likelihood comes from the probability that your data matches your model (*i.e. the probability of the data given the model*).

Let's talk about this in the context of fitting 2-D data to a curve.

```python
from IPython.display import Image
Image("http://www.astro.columbia.edu/~lia/astroML_plots/observed_halo.png", width=800)
```

Sure, it might look like a line.  But that's not a physically motivated model.  The data is the surface brightness profile of an X-ray scattering halo.  The free parameters I'll use are the maximum dust grain size ($a$), the power law exponent to the grain size distribution ($p$), and the ISM column ($N_H$).

In the classical approach, we **often assume that our measurements take on a Gaussian distribution**, with the mean value ($\mu$) described by our model (the curve) and the standard deviation described by our error bar.  So if we made the same measurement over and over again, we'd get a nice little Gaussian distribution centered on our model.

The probability of our data point being consistent with the model is described by that Gaussian. <br>
(*The Gaussian describes the probability of getting that data point measurement, given our model.*)

**Assuming that our data points are not correlated**, we multiply these probabilities to compute the likelihood.  This assumption is very important, and if it's not true you need to use covariance matrices or something.  I have never needed to use them, so unfortunately I cannot give you a very comprehensive overview.

Pretending for now that you have <b>uncorrelated Gaussian data points</b>, the likelihood function for your model is:

$$ L \propto \Pi_{i=1}^N \exp \left( \frac{-(x_i - \mu)^2}{2 \sigma_i^2} \right) $$

+ Data point: $x_i$
+ Error bar: $\sigma_i$
+ Mean (model) value: $\mu$
+ Total number of data points: $N$

** Your Goal: ** Find the model that maximizes the likelihood!

** A Convenient Trick: ** Take a logarithm, so we can change the product into a sum.

$$ \ln(L) = {\rm const} - \frac{1}{2} \Sigma_{i=1}^N \frac{(x_i - \mu)^2}{\sigma_i^2} $$

To maximize this value, we just have to minimize the second term.  You might recognize this as the method of minimizing chi-squared, which many of us know and love!

Recall that the chi-squared many of us know and love is $\chi^2 \equiv \Sigma_{i=1}^N (x_i-\mu)^2 / \sigma_i^2$

Once we've minimized $\chi^2$, we've found the best model curve for our data.  Frequentists rejoice!

Here's what happened when I applied <code>scipy.optimize.leastsq</code> to my data and model:

+ The grey dashed line was just my starting point.
+ The horizontal lines are not error bars, they are the width of my radial profile bins.
+ The red line shows the best fit curve: $a = 0.13$ um, $p=3.8$, and $N_H = 3 \times 10^{22}$ cm$^{-2}$.
+ ** The reduced chi-squared (or $\chi^2_{dof}$) is 0.4 **  There will be more on this later.

## Bet you didn't know you've been Maximizing Likelihood all this time.

### Great, so I have some data. How do I find the Universal law (distribution) that produced it?

Suppose you have a set of photometry for O type stars, and you want to know the absolute magnitude of O stars set by nature.  You might argue that their absolute magnitudes are described by a Gaussian distribution $N(\mu,\sigma)$.

**Case 1:** If your photometry measurements are **homoscedastic**, you can use #AstroML Equation 4.4 and 4.5

$$ \mu^0 = \frac{1}{N} \Sigma_{i=1}^N x_i $$

**Case 2:** If your photometry measurements have large error bars and are **heteroscedastic**, you should use #AstroML Equation 4.9, which is a weighted average.

$$ \mu^0 = \frac{\Sigma x_i \sigma_i^{-2}}{\Sigma \sigma_i^{-2}} $$

Either way, these formulae were arrived at by setting $\frac{d ln(L)}{d\mu} = 0$. 

**Taking an average is the statistical equivalent of maximizing the likelihood for Gaussian data!!**

In statistical speak:

+ An **Estimator** is a metric used to make inferences about real life from your data.<br>
The average $\mu^0$ was arrived at by maximizing the likelihood, thus it is a **Maximum Likelihood Estimator**

+ An estimator has an associated uncertainty, the **MLE Confidence Interval** or **uncertainty of the mean**.  #AstroML Equations 4.7 and 4.10 show how this is either the standard error of the mean (case 1):
$$ \sigma_\mu = \frac{\sigma}{\sqrt N} $$
or the propogated error (case 2):
$$ \sigma_\mu = \left( \Sigma_{i=1}^N \sigma_i^2 \right)^{1/2}$$

** Doing something as simple as averaging your data is deeply rooted in probabilistic theory. **

I realized that I could think of a physically motivated model of three screens, each with their own dust grain size distribution, for a total of 12 parameters.  But I was hesitant to go to such lengths.  I posted the following question on twitter:

** "Fit a physically motivated model with 12 parameters to 16 data points, or choose the uninteresting 3 parameter model?"**

To my surprise, this stirred some debate that strongly illustrated the differences between Bayesian and frequentist mindsets.

** Frequentist mindset **

+ "Always go with the minimum number of free parameters. You can fit anything with enough knobs to twiddle."
+ Our fits are "too good"
+ "Overfitting"

** Bayesian mindset **

+ "Far more honest to use a general model + informative priors that actually represent your beliefs."
+ "Choosing the simpler model is the same as using infinitely informative priors."
+ "Wait. That's not a thing."

There is value in all of these comments.

It is true that we can get an increasingly good fit with a large number of model parameters.  **Just because we can get $\chi^2$ to be arbitrarily small, doesn't mean we should.**

This leads to the frequentist rule-of-thumb (dogma) that the reduced chi-squared ($\bar{\chi}^2$, $\chi^2_{dof}$, or the "chi-squared per degrees of freedom") should be close to 1.  

** But why should $\chi^2_{dof} = 1$?  Using MLE, we want $\chi^2$ to get as close to zero as possible, right?**

If we were to repeat the experiment many times, evaluating $\ln (L)$ over and over again, it would take on the $\chi^2$ distribution, which has an expectation value of $(N-k)$ i.e. the "degrees of freedom."

As $N-k$ gets large, the $\chi^2$ distribution approaches Gaussian.

** So when we evaluate our models based on $\chi^2_{dof}$, we are making an assumption about the probability distribution of $\ln (L)$. **

###My new favorite rule-of-thumb:

+ Use $\chi^2_{dof}$ to judge your error bars.
+ A residual plot should be used to interpret your model fit.

```python
# Author: Jake VanderPlas
# License: BSD
#   The figure produced by this code is published in the textbook
#   "Statistics, Data Mining, and Machine Learning in Astronomy" (2013)
#   For more information, see http://astroML.github.com
#   To report a bug or issue, use the following forum:
#    https://groups.google.com/forum/#!forum/astroml-general
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt

#----------------------------------------------------------------------
# This function adjusts matplotlib settings for a uniform feel in the textbook.
# Note that with usetex=True, fonts are rendered with LaTeX.  This may
# result in an error if LaTeX is not installed on your system.  In that case,
# you can set usetex to False.
from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=8, usetex=True)

#------------------------------------------------------------
# Generate Dataset
np.random.seed(1)

N = 50
L0 = 10
dL = 0.2

t = np.linspace(0, 1, N)
L_obs = np.random.normal(L0, dL, N)

#------------------------------------------------------------
# Plot the results
fig = plt.figure(figsize=(15, 15))
fig.subplots_adjust(left=0.1, right=0.95, wspace=0.05,
                    bottom=0.1, top=0.95, hspace=0.05)

y_vals = [L_obs, L_obs, L_obs, L_obs + 0.5 - t ** 2]
y_errs = [dL, dL * 2, dL / 2, dL]
titles = ['correct errors',
          'overestimated errors',
          'underestimated errors',
          'incorrect model']

for i in range(4):
    ax = fig.add_subplot(2, 2, 1 + i, xticks=[])

    # compute the mean and the chi^2/dof
    mu = np.mean(y_vals[i])
    z = (y_vals[i] - mu) / y_errs[i]
    chi2 = np.sum(z ** 2)
    chi2dof = chi2 / (N - 1)

    # compute the standard deviations of chi^2/dof
    sigma = np.sqrt(2. / (N - 1))
    nsig = (chi2dof - 1) / sigma

    # plot the points with errorbars
    ax.errorbar(t, y_vals[i], y_errs[i], fmt='.k', ecolor='gray', lw=1)
    ax.plot([-0.1, 1.3], [L0, L0], ':k', lw=1)

    # Add labels and text
    ax.text(0.95, 0.95, titles[i], ha='right', va='top',
            transform=ax.transAxes,
            bbox=dict(boxstyle='round', fc='w', ec='k'))
    ax.text(0.02, 0.02, r'$\hat{\mu} = %.2f$' % mu, ha='left', va='bottom',
            transform=ax.transAxes)
    ax.text(0.98, 0.02,
            r'$\chi^2_{\rm dof} = %.2f\, (%.2g\,\sigma)$' % (chi2dof, nsig),
            ha='right', va='bottom', transform=ax.transAxes)

    # set axis limits
    ax.set_xlim(-0.05, 1.05)
    ax.set_ylim(8.6, 11.4)

    # set ticks and labels
    ax.yaxis.set_major_locator(plt.MultipleLocator(1))

    if i > 1:
        ax.set_xlabel('observations')

    if i % 2 == 0:
        ax.set_ylabel('Luminosity')
    else:
        ax.yaxis.set_major_formatter(plt.NullFormatter())

plt.show()
```

### 4.3.2: If all else fails, use a "scoring system" to "take into account the model complexity and penalize models for additional parameters not supported by data."$^4$

Here's where #AstroML introduces the **Aikake information criterion (AIC)**

$$ AIC \equiv -2 \ln(L^0(M)) + 2 k + \frac{2 k (k+1)}{N-k-1} $$

I haven't read the paper, so this equation is "magic" for now.

** The smaller the AIC, the better the score for your model. **

For two models with the same $L^0(M)$, the award goes to which ever model has a smaller $k$.

# 4.4: The Expectation Maximization Algorithm

This section covers the "very special case of a fairly complex likelihood which can still be maximized using a relatively simple and straightforward numerical method."

It is the case of measurements that are clearly non-Gaussian (Figure 4.2), fit as a sum of Guassian components.  

Since I don't have much experience to add to this particular discussion, I recommend that interested parties try implementing the EM algorithm for one of their hack days.

```python
# Author: Jake VanderPlas
# License: BSD
#   The figure produced by this code is published in the textbook
#   "Statistics, Data Mining, and Machine Learning in Astronomy" (2013)
#   For more information, see http://astroML.github.com
#   To report a bug or issue, use the following forum:
#    https://groups.google.com/forum/#!forum/astroml-general
from matplotlib import pyplot as plt
import numpy as np
from sklearn.mixture import GMM

#----------------------------------------------------------------------
# This function adjusts matplotlib settings for a uniform feel in the textbook.
# Note that with usetex=True, fonts are rendered with LaTeX.  This may
# result in an error if LaTeX is not installed on your system.  In that case,
# you can set usetex to False.
from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=8, usetex=True)

#------------------------------------------------------------
# Set up the dataset.
#  We'll use scikit-learn's Gaussian Mixture Model to sample
#  data from a mixture of Gaussians.  The usual way of using
#  this involves fitting the mixture to data: we'll see that
#  below.  Here we'll set the internal means, covariances,
#  and weights by-hand.
np.random.seed(1)

gmm = GMM(3, n_iter=1)
gmm.means_ = np.array([[-1], [0], [3]])
gmm.covars_ = np.array([[1.5], [1], [0.5]]) ** 2
gmm.weights_ = np.array([0.3, 0.5, 0.2])

X = gmm.sample(1000)

#------------------------------------------------------------
# Learn the best-fit GMM models
#  Here we'll use GMM in the standard way: the fit() method
#  uses an Expectation-Maximization approach to find the best
#  mixture of Gaussians for the data

# fit models with 1-10 components
N = np.arange(1, 11)
models = [None for i in range(len(N))]

for i in range(len(N)):
    models[i] = GMM(N[i]).fit(X)

# compute the AIC and the BIC
AIC = [m.aic(X) for m in models]
BIC = [m.bic(X) for m in models]

#------------------------------------------------------------
# Plot the results
#  We'll use three panels:
#   1) data + best-fit mixture
#   2) AIC and BIC vs number of components
#   3) probability that a point came from each component

fig = plt.figure(figsize=(5*3, 1.7*3))
fig.subplots_adjust(left=0.12, right=0.97,
                    bottom=0.21, top=0.9, wspace=0.5)


# plot 1: data + best-fit mixture
ax = fig.add_subplot(131)
M_best = models[np.argmin(AIC)]

x = np.linspace(-6, 6, 1000)
logprob, responsibilities = M_best.eval(x)
pdf = np.exp(logprob)
pdf_individual = responsibilities * pdf[:, np.newaxis]

ax.hist(X, 30, normed=True, histtype='stepfilled', alpha=0.4)
ax.plot(x, pdf, '-k')
ax.plot(x, pdf_individual, '--k')
ax.text(0.04, 0.96, "Best-fit Mixture",
        ha='left', va='top', transform=ax.transAxes)
ax.set_xlabel('$x$')
ax.set_ylabel('$p(x)$')


# plot 2: AIC and BIC
ax = fig.add_subplot(132)
ax.plot(N, AIC, '-k', label='AIC')
ax.plot(N, BIC, '--k', label='BIC')
ax.set_xlabel('n. components')
ax.set_ylabel('information criterion')
ax.legend(loc=2)


# plot 3: posterior probabilities for each component
ax = fig.add_subplot(133)

p = M_best.predict_proba(x)
p = p[:, (1, 0, 2)]  # rearrange order so the plot looks better
p = p.cumsum(1).T

ax.fill_between(x, 0, p[0], color='gray', alpha=0.3)
ax.fill_between(x, p[0], p[1], color='gray', alpha=0.5)
ax.fill_between(x, p[1], 1, color='gray', alpha=0.7)
ax.set_xlim(-6, 6)
ax.set_ylim(0, 1)
ax.set_xlabel('$x$')
ax.set_ylabel(r'$p({\rm class}|x)$')

ax.text(-5, 0.3, 'class 1', rotation='vertical')
ax.text(0, 0.5, 'class 2', rotation='vertical')
ax.text(3, 0.3, 'class 3', rotation='vertical')

plt.show()
```
